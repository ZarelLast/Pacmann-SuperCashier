```python
from modul_only_projek_i import Transaction
from tabulate import tabulate
import os
```

```python
def input_validasi(teks, type_data):
  """
  Validasi Input dari inputan pelanggan

    Parameter :
    -----------
    teks : str
      teks berisikan teks yang akan di keluarkan pada saat input
    type_data : type
      type_data berisikan type data yang akan digunakan untuk input

    Return :
    -----------
    value : str/int
      value berisikan inputan user yang sudah di validasi
      jika inputan error maka akan return 0 atau string kosong

  """
  if type_data == str:
    try:
      value = str(input(f"{teks}: "))
    except ValueError:
      value = ''
      print(ValueError(f"Input Error - {teks} tidak sesuai, harap update {teks}"))

  if type_data == int:
    try:
      value = int(input(f"{teks}: "))
    except ValueError:
      value = 0
      print(ValueError(f"Input Error - {teks} bukan angka, harap update {teks}"))

  return value
```

```python
def test_case_output(data):
  """
  Untuk membuat output sesuai dengan keinginan test case

    Parameter :
    -----------
    data : list
      data berisikan list yang didalamnya terdapat dictionary
      [{'nama':'soto', 'jumlah':10, 'harga':8000}]

  """
  res = {}
  for barang in data:
    res[barang['nama']] = [barang['jumlah'], barang['harga']]
  print(f"Barang yang dibeli: \n{res}")
```

```python
barang = Transaction()
barang.reset_transaction()

stats = True
while stats:
  os.system('cls')
  print("\n\nMenu Utama")
  header = ["No.", "Opsi"]
  table = [
      [0, "Check Keranjang"],
      [1, "Tambah Item"],
      [2, "Update Nama Item"],
      [3, "Update Jumlah Item"],
      [4, "Update Harga Item"],
      [5, "Hapus Item"],
      [6, "Bayar dan Checkout"],
      [7, "Batal Belanja"]
    ]
  print(tabulate(table, headers = header, tablefmt = "github"))

  try:
    opsi = int(input("\nPilih opsi: "))
  except ValueError:
    opsi = 10

  if opsi == 0:
    barang.check_order()
  elif opsi == 1:
    # nama_item = input("Nama barang: ")
    # jumlah_item = input("Jumlah barang: ")
    # harga_item = int(input("Harga barang: "))
    nama_item = input_validasi(teks="Nama Barang", type_data=str)
    jumlah_item = input_validasi(teks="Jumlah Barang", type_data=int)
    harga_item = input_validasi(teks="Harga Barang", type_data=int)
    barang.add_items([nama_item, jumlah_item, harga_item])
  elif opsi == 2:
    if len(barang.keranjang) != 0:
      try:
        id_item = int(input("Id barang: "))
      except:
        print('id tidak tersedia')
        continue
      nama_baru_item = input_validasi(teks="Nama Barang", type_data=str)
      barang.update_item_name(id_item, nama_baru_item)
    else:
      print('Keranjang anda masih kosong!')
  elif opsi == 3:
    if len(barang.keranjang) != 0:
      try:
        id_item = int(input("Id barang: "))
      except:
        print('id tidak tersedia')
        continue
      jumlah_baru_item = input_validasi(teks="Jumlah Barang", type_data=int)
      barang.update_item_qty(id_item, jumlah_baru_item)
    else:
      print('Keranjang anda masih kosong!')
  elif opsi == 4:
    if len(barang.keranjang) != 0:
      try:
        id_item = int(input("Id barang: "))
      except:
        print('id tidak tersedia')
        continue
      harga_baru_item = input_validasi(teks="Harga Barang", type_data=int)
      barang.update_item_price(id_item, harga_baru_item)
    else:
      print('Keranjang anda masih kosong!')
  elif opsi == 5:
    if len(barang.keranjang) != 0:
      try:
        id_item = int(input("Id barang: "))
      except:
        print('id tidak tersedia')
        continue
      pilih = input("Anda yakin ingin menghapus item? (y=ya, t=tidak): ")
      if pilih.lower() == 'y' or pilih.lower() == 'ya':
        barang.delete_item(id_item)
      elif pilih.lower() == 't' or pilih.lower() == 'tidak':
        pass
      else:
        print('Opsi tidak tersedia!')
    else:
      print('Keranjang anda masih kosong!')
  elif opsi == 6:
    if len(barang.keranjang) != 0:
      barang.check_order()

      pilih = input("anda yakin ingin checkout? (y=ya, t=tidak): ")
      if pilih.lower() == 'y' or pilih.lower() == 'ya':
        test_case_output(barang.keranjang)
        barang.total_price()
        barang.reset_transaction()
        stats = False
      elif pilih.lower() == 't' or pilih.lower() == 'tidak':
        pass
      else:
        print('Opsi tidak tersedia!')
    else:
      print('Keranjang anda masih kosong!')
  elif opsi == 7:
    pilih = input("anda yakin ingin Batal Beli? (y=ya, t=tidak): ")
    if pilih.lower() == 'y' or pilih.lower() == 'ya':
      barang.reset_transaction()
      stats = False
    elif pilih.lower() == 't' or pilih.lower() == 'tidak':
      pass
    else:
      print('Opsi tidak tersedia!')
  else:
    print("Opsi tidak tersedia!")
```
